# Matplotlib Module 

# Basic Plotting

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 25, 30, 35, 45, 70]
plt.plot(x, y)
plt.title("Basic line Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 25, 30, 35, 45, 70]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

# Bar Chart

```python
x = ['Apple', 'Banana', 'Cherry']
y = [40, 70, 30]
plt.bar(x, y, color='orange')
plt.title("Fruit Bar Chart")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

# Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 35, 15, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Language Popularity")
plt.axis('equal')  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
group1 = [20, 34, 30, 35]
group2 = [25, 32, 34, 20]

x = np.arange(len(categories))
width = 0.35 
plt.bar(x - width/2, group1, width, label='Group 1', color='skyblue')
plt.bar(x + width/2, group2, width, label='Group 2', color='orange')

plt.xlabel('Categories')
plt.ylabel('Scores')
plt.title('Grouped Bar Chart Example')
plt.xticks(x, categories)
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 12, 20]

plt.plot(x, y, marker='*', color='purple', linestyle='-', linewidth=2, markersize=12, label="Star Line")

plt.title("Line Graph with Star Markers")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

points = 5
angles = np.linspace(0, 2 * np.pi, points * 2 + 1)
radius = np.tile([1, 0.4], points)
radius = np.append(radius, radius[0])
x = radius * np.cos(angles)
y = radius * np.sin(angles)

plt.figure(figsize=(6, 6))
plt.plot(x, y, color='crimson', linewidth=2)
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D']
cat1 = np.array([30, 50, 20, 40])
cat2 = np.array([70, 50, 80, 60])

total = cat1 + cat2
cat1 = cat1 / total * 100
cat2 = cat2 / total * 100

x = [0, 1, 2, 3]

plt.bar(x, cat1)
plt.bar(x, cat2, bottom=cat1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['A', 'B', 'C', 'D']
y = [20, 35, 30, 35]
errors = [2, 3, 4, 1]

x_pos = np.arange(len(x))

plt.bar(x_pos, y, yerr=errors)
plt.xticks(x_pos, x)
plt.show()
```
