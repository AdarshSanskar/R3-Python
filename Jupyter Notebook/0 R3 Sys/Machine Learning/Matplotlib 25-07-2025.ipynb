# Matplotlib Module 

# Basic Plotting

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 25, 30, 35, 45, 70]
plt.plot(x, y)
plt.title("Basic line Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 25, 30, 35, 45, 70]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

# Bar Chart

```python
x = ['Apple', 'Banana', 'Cherry']
y = [40, 70, 30]
plt.bar(x, y, color='orange')
plt.title("Fruit Bar Chart")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

# Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 35, 15, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Language Popularity")
plt.axis('equal')  
plt.show()
```
